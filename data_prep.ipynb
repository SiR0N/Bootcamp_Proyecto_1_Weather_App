```python
codigos_estaciones = {
    "102": "J.M.D. Moratalaz",
    "103": "J.M.D. Villaverde",
    "104": "E.D.A.R. La China",
    "106": "Centro Mpal. De Acústica",
    "107": "J.M.D. Hortaleza",
    "108": "Peñagrande",
    "109": "J.M.D. Chamberí",
    "110": "J.M.D. Centro",
    "111": "J.M.D. Chamartín",
    "112": "J.M.D. Vallecas 1",
    "113": "J.M.D. Vallecas 2",
    "114": "Matadero 01",
    "115": "Matadero 02",
    "004": "Plaza España",
    "008": "Escuelas Aguirre",
    "016": "Arturo Soria",
    "018": "Farolillo",
    "024": "Casa de Campo",
    "035": "Plaza del Carmen",
    "036": "Moratalaz",
    "038": "Cuatro Caminos",
    "039": "Barrio del Pilar",
    "054": "Ensanche de Vallecas",
    "056": "Plaza Elíptica",
    "058": "El Pardo",
    "059": "Juan Carlos I"
}
```

```python
codigos_magnitudes = {
    80: {"parametro": "Radiación ultravioleta", "unidad": "mW/m2"},
    81: {"parametro": "Velocidad del viento", "unidad": "m/s"},
    82: {"parametro": "Dirección del viento", "unidad": "grados"},
    83: {"parametro": "Temperatura", "unidad": "ºC"},
    86: {"parametro": "Humedad relativa", "unidad": "%"},
    87: {"parametro": "Presión barométrica", "unidad": "mb"},
    88: {"parametro": "Radiación solar", "unidad": "W/m2"},
    89: {"parametro": "Precipitación", "unidad": "l/m2"}
}
```

```python
df["ESTACION_NOMBRE"] = df["ESTACION"].astype(str).map(codigos_estaciones)
df["MAGNITUD_NOMBRE"] = df["MAGNITUD"].map(lambda x: codigos_magnitudes[x]["parametro"])
df["UNIDAD"] = df["MAGNITUD"].map(lambda x: codigos_magnitudes[x]["unidad"])
```

```python
df.head()
```

```python
rows = []

for _, row in df.iterrows():
    for h in range(1, 25):
        valor = row[f"H{h:02d}"]
        valid = row[f"V{h:02d}"]
        
        fecha = pd.Timestamp(row["ANO"], row["MES"], row["DIA"]) + pd.Timedelta(hours=h-1)

        rows.append({
            "PROVINCIA": row["PROVINCIA"],
            "MUNICIPIO": row["MUNICIPIO"],
            "ESTACION": row["ESTACION"],
            "ESTACION_NOMBRE": row["ESTACION_NOMBRE"],
            "MAGNITUD": row["MAGNITUD"],
            "MAGNITUD_NOMBRE": row["MAGNITUD_NOMBRE"],
            "UNIDAD": row["UNIDAD"],
            "ANO": row["ANO"],
            "MES": row["MES"],
            "DIA": row["DIA"],
            "HORA": h,
            "VALOR": valor,
            "VALIDACION": valid,
            "FECHA_HORA": fecha

        })

df_long = pd.DataFrame(rows)
```

```python
df_long
```

```python
df_valores = df_long.pivot_table(
    index=["PROVINCIA", "MUNICIPIO", "ESTACION", "ESTACION_NOMBRE", "FECHA_HORA"],
    columns="MAGNITUD_NOMBRE",
    values="VALOR",
    aggfunc="first"
)
```

```python
df_valid = df_long.pivot_table(
    index=["PROVINCIA", "MUNICIPIO", "ESTACION", "ESTACION_NOMBRE", "FECHA_HORA"],
    columns="MAGNITUD_NOMBRE",
    values="VALIDACION",
    aggfunc="first"
)
```

```python
df_valid = df_valid.add_suffix("_valid")
```

```python
df_wide = pd.concat([df_valores, df_valid], axis=1).reset_index()
```

```python
df_wide.to_csv("historico_expandido.csv", index=False, sep=";")
```
