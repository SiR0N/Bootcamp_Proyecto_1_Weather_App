```python
import pandas as pd
import hashlib
import json
```

```python
users = [
    {
        "name": "Jose",
        "pass": "1234Jose?"
    },
    {
        "name": "Laura",
        "pass": "1234Laura?"
    }
]
```

```python
dic = []
for user in users:
    #print(user["name"])
    #print(user["pass"])
    pas = hashlib.sha256(user["pass"].encode()).hexdigest()
    dic.append(
        {
            "name": user["name"],
            "pass": pas
        }
    )
    
    #print(pas)
```

```python
dic
```

```python
json_data = json.dumps(dic, indent=4)
```

```python
json_data
```

```python
with open("users.json", "w") as f:
    json.dump(dic, f, indent=4)
```

```python
with open("users.json", "r") as f:
    data = json.load(f)
```
